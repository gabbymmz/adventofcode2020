```python
input_2 = open("Day2_input.txt", 'r')

passwords_i = []
for row in input_2:
    stripped_line = row.strip()
    split_line = stripped_line.split()
    passwords_i.append(split_line)

#passwords_i
for row in passwords_i:
    letter = row[1].replace(":","")
    row[1] = letter
passwords_i
```

```python
list_test = [
    ['1-3', 'a', 'abcde'],
    ['1-3', 'b', 'cdefg'],
    ['2-9', 'c', 'ccccccccc']
]
```

```python
for row in list_test:
    my_str = row[0]
#def range_to_nums(my_str):
    res = sum(((list(range(*[int(b) + c  
        for c, b in enumerate(a.split('-'))]))
        if '-' in a else [int(a)]) for a in my_str.split(', ')), [])
    print (res)
```

```python
#pull min & max rules
min_max = []
for row in passwords_i:
    my_str = row[0]
    res = sum(((list(range(*[int(b) + c  
        for c, b in enumerate(a.split('-'))]))
        if '-' in a else [int(a)]) for a in my_str.split(', ')), [])
    range_to_num = [res[0], res[-1]]
    #new_row = row.append(range_to_num)
    min_max.append(range_to_num)

passwords = [a + b for a, b in zip(passwords_i, min_max)]
passwords
```

```python
for row in passwords:        
    word = row[2]
    letter = row[1]
    letter_count = word.count(letter)
    min_v = row[3]
    max_v = row[4]
    if letter_count >= min_v and letter_count <= max_v:
        row.append('Valid')
    else:
        row.append('Not Valid')
```

```python
count_valid=0
count_nonval=0
for row in passwords:
    if row[5] == 'Valid':
        count_valid += 1
    else:
        count_nonval += 1
print(count_valid)
print (count_valid + count_nonval == len(passwords))
```

```python
#Part 2

# s = 'shak#spea#e'
# c = '#'
# print([pos for pos, char in enumerate(s) if char == c])

for row in passwords:
    word = row[2]
    letter = row[1]
    check= [pos for pos, char in enumerate(word) if char == letter]
    #print (check)
    positions = [x+1 for x in check]
    #print (positions)
    if (row[3] in positions and row [4] not in positions) or (row[3] not in positions and row[4] in positions):
        row.append("Valid again")
    else:
        row.append("Invalid")
```

```python
count_val_again=0
count_non_again=0
for row in passwords:
    if row[6] == 'Valid again':
        count_val_again += 1
    else:
        count_non_again += 1
print(count_val_again)
print (count_val_again + count_non_again == len(passwords))
```
